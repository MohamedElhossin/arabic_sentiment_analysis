```
import pandas as pd 
import numpy as np 

import plotly.express as px
import plotly.graph_objects as go

pd.set_option('max_colwidth', 200)
```

```
df = pd.read_csv('/content/drive/My Drive/data/unbalanced-reviews.csv', sep='\t', encoding='utf-16')
```

```
df.sample(20)
```

```
# Rename columns name
df.rename(columns={'no' : 'user_id', 'Hotel name' : 'hotel_name',
                   'user type' : 'user_type', 'room type': 'room_type'}, 
                    inplace=True)
```

```
# Show information about datasets 

print('Number of reviews {}'.format(df.shape[0]))
print('Number of hotels {}'.format(len(df['hotel_name'].unique())))
print('Number of postive reviews {}'.format(len(df[df['rating'] > 3])))
print('Number of negative reviews {}'.format(len(df[df['rating'] < 3])))
print('Number of netural reviews {}'.format(len(df[df['rating'] == 3])))
```

```
# filter dataset for top 10 hotels which take most numbers of  reviews 
hotel_count = df['hotel_name'].value_counts()
hotel_name = hotel_count[0:10].index

df_top10 = df[df['hotel_name'].isin(hotel_name)]

print('Number of reviews {}'.format(df_top10.shape[0]))
print('Number of hotels {}'.format(len(df_top10['hotel_name'].unique())))
print('Number of postive reviews {}'.format(len(df_top10[df_top10['rating'] > 3])))
print('Number of negative reviews {}'.format(len(df_top10[df_top10['rating'] < 3])))
print('Number of natural reviews {}'.format(len(df_top10[df_top10['rating'] == 3])))
```

#**Preprocessing data** 

```
# remeove 'فندق' , 'مكة' and '-' from hotel name 
# remove space from hotel name

df_top10['hotel_name'] = df_top10['hotel_name'].replace(u'فندق أبراج الصفوة – دار الغفران' , 'أبراج الصفوة')
df_top10['hotel_name'] = df_top10['hotel_name'].str.replace(u'فندق' , '')
df_top10['hotel_name'] = df_top10['hotel_name'].str.replace(u'مكة' , '')
df_top10['hotel_name'] = df_top10['hotel_name'].str.replace(u' -' , '')
df_top10['hotel_name'] = df_top10['hotel_name'].str.strip()
```

```
# get number of nights and change to int type

df_top10['nights'] = df_top10['nights'].str.replace(u'أقمت ليلة واحدة', '1')
df_top10['nights'] = df_top10['nights'].str.replace(u'أقمت ليلتين', '2')
df_top10['nights'] = df_top10['nights'].str.replace(r'\D', '')
df_top10['nights'] = df_top10['nights'].str.replace('', '1')

df_top10['nights'] = df_top10['nights'].astype('int')
```

```
# assign 1 for postive review, 
# 0 for negative and 
# 2  for netural review

df_top10.loc[df_top10['rating'] > 3, 'sentiment'] = 'ايجابي'
df_top10.loc[df_top10['rating'] < 3, 'sentiment'] = 'سلبي'
df_top10.loc[df_top10['rating'] == 3, 'sentiment'] = 'طبيعي'
```

#**Visualization**

```
# percentage of positive reviews

user_number = df_top10.groupby('hotel_name')['user_id'].nunique().reset_index(name = 'user_number')
reviews_number = df_top10.groupby(['hotel_name', 'sentiment'])['review'].count().reset_index(name = 'review_number')

avg_reviews_number = reviews_number.merge(user_number , on = 'hotel_name', how='left')
avg_reviews_number['review %'] = round(avg_reviews_number['review_number'] / avg_reviews_number['user_number'], 2) * 100

postive_reviews = avg_reviews_number[avg_reviews_number['sentiment'] == u'ايجابي']
postive_reviews.sort_values(by='review %', ascending=False, inplace=True)
```

```
colors = ['green', 'blue', 'red']
hotels = df_top10['hotel_name'].unique()
fig = go.Figure()

for hotel in hotels:
    temp = list(avg_reviews_number[avg_reviews_number['hotel_name'] == hotel]['review_number'])

    fig.add_trace(go.Pie(labels=[' أجابي',' طبيعي',' سلبي'],
                             values=[temp[0],temp[2],temp[1]],
                              opacity = 0.8,
                              hole=.3,
                              visible = False),
                   )
    
    fig.update_traces(hoverinfo='label+percent', textinfo='percent', textfont_size=20,
                  marker=dict(colors=colors))
    
    
  

   
fig.update_layout(
       updatemenus=[
                    dict(
                        buttons=list([
                                      dict(label=hotels[0],
                                           method='update',
                                           args=[{'visible': 
                                                  [True,False,False,False,False,
                                                   False,False,False,False,False]}
                                                 ]),
                                      dict(label=hotels[1],
                                           method='update',
                                           args=[{'visible': 
                                                  [False,True,False,False,False,
                                                   False,False,False,False,False]}
                                                 ]),
                                      dict(label=hotels[2],
                                           method='update',
                                           args=[{'visible': 
                                                  [False,False,True,False,False,
                                                   False,False,False,False,False]}
                                                 ]),
                                      dict(label=hotels[3],
                                           method='update',
                                           args=[{'visible': 
                                                  [False,False,False,True,False,
                                                   False,False,False,False,False]}
                                                 ]),
                                      dict(label=hotels[4],
                                           method='update',
                                           args=[{'visible': 
                                                  [False,False,False,False,True,
                                                  False,False,False,False,False]}
                                                 ]),
                                      dict(label=hotels[5],
                                           method='update',
                                           args=[{'visible': 
                                                  [False,False,False,False,False,
                                                   True,False,False,False,False]}
                                                 ]),
                                      dict(label=hotels[6],
                                           method='update',
                                           args=[{'visible': 
                                                  [False,False,False,False,False,
                                                   False,True,False,False,False]}
                                                 ]),
                                      dict(label=hotels[7],
                                           method='update',
                                           args=[{'visible': 
                                                  [False,False,False,False,False,
                                                   False,False,True,False,False]}
                                                ]),
                                     dict(label=hotels[8],
                                           method='update',
                                           args=[{'visible': 
                                                  [False,False,False,False,False,
                                                   False,False,False,True,False]}
                                                ]),
                                      dict(label=hotels[9],
                                           method='update',
                                           args=[{'visible': 
                                                  [False,False,False,False,False,
                                                   False,False,False,False,True]}
                                                ]),                    
                        ]),
                        direction="down",
                        #pad={"r": 10, "t": 10},
                        showactive=True,
                        x=0.002,
                        xanchor="left",
                        y=1.14,
                        yanchor="top" 
                    )
       ]
   )

fig.update_layout(
    title={
        'y':0.9,
        'x':0.6,
        'xanchor': 'right',
        'yanchor': 'top'})

fig.show()
  
```

```

fig = px.bar(postive_reviews, x ='hotel_name', y = 'review %',
       title = 'نسبة المراجعات الإيجابية لكل فندق',
       color = 'user_number',
       hover_data = ['hotel_name', 'review_number', 'user_number'],
       labels = {'hotel_name' : 'o اسم الفندق' , 'review %' : 'o النسبة المؤويه', 
                'user_number' : 'o اعداد النزلاء', 'review_number' : 'o أعداد المرجعات الاجابيه'})

fig.update_traces(texttemplate='%{y:.2s}%', textposition='outside')

fig.update_layout(
    title={
        'y':0.9,
        'x':0.6,
        'xanchor': 'right',
        'yanchor': 'top'})

fig.show()
```

```
# avg nights stayed by user

number_of_nights = df_top10.groupby('hotel_name')['nights'].sum().reset_index()

number_of_nights = number_of_nights.merge(user_number , on = 'hotel_name', how='left')
number_of_nights['avg'] = round(number_of_nights['nights'] / number_of_nights['user_number']) 

number_of_nights.sort_values(by='avg', ascending=False, inplace=True)
```

```
fig = px.bar(number_of_nights, x ='hotel_name', y = 'avg',
             title = 'متوسط عدد الليالي التي يقيمها الفرد ',
             hover_data = ['hotel_name', 'avg', 'user_number'],
             labels = {'hotel_name' : 'o اسم الفندق' , 'avg' : 'o عدد الليالي', 
                       'user_number' : 'o اعداد النزلاء'},
             opacity = 0.8)

fig.update_traces(texttemplate='%{y}', textposition='outside')

fig.update_layout(
    title={
        'y':0.9,
        'x':0.6,
        'xanchor': 'right',
        'yanchor': 'top'})

fig.show()
```
